```python
import gymnasium as gym
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.multiprocessing as mp
import numpy as np
import matplotlib.pyplot as plt
```

```python
ENV_ID        = "CartPole-v1"
NUM_WORKERS   = 4
MAX_EPISODES  = 2500
N_STEPS       = 25
GAMMA         = 0.99
LR            = 1e-3
SMOOTHING     = 50
EVAL_EPISODES = 10
```

```python
class ActorCritic(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.fc = nn.Linear(input_dim, 128)
        self.pi = nn.Linear(128, output_dim)
        self.v  = nn.Linear(128, 1)
    def forward(self, x):
        x = F.relu(self.fc(x))
        return self.pi(x), self.v(x)
```

```python
class A3CWorker(mp.Process):
    def __init__(self, wid, env_id, model, opt, counter, queue):
        super().__init__()
        self.wid     = wid
        self.env_id  = env_id
        self.model   = model
        self.opt     = opt
        self.counter = counter
        self.queue   = queue

    def run(self):
        env    = gym.make(self.env_id)
        state, _ = env.reset()
        ep_reward = 0

        while not self._stop_condition():
            trajectory, state, ep_reward = self._collect_trajectory(env, state, ep_reward)
            loss = self._compute_loss(trajectory)
            self._apply_gradients(loss)

        env.close()

    def _stop_condition(self):
        with self.counter.get_lock():
            return self.counter.value >= MAX_EPISODES

    def _collect_trajectory(self, env, state, ep_reward):
        log_probs, values, rewards = [], [], []
        done = False

        for _ in range(N_STEPS):
            s_tensor = torch.as_tensor(state, dtype=torch.float32)
            logits, value = self.model(s_tensor)
            dist = torch.distributions.Categorical(logits.softmax(-1))
            action = dist.sample().item()

            next_state, r, term, trunc, _ = env.step(action)
            done = term or trunc

            log_probs.append(dist.log_prob(torch.tensor(action)))
            values.append(value.squeeze(0))
            rewards.append(r)
            ep_reward += r
            state = next_state

            if done:
                with self.counter.get_lock():
                    self.counter.value += 1
                    self.queue.put(("[reward]", ep_reward))
                ep_reward, state = 0, env.reset()[0]
                break

        trajectory = {
            "log_probs": log_probs,
            "values":    values,
            "rewards":   rewards,
            "done":      done,
            "next_state": state
        }
        return trajectory, state, ep_reward

    def _compute_loss(self, traj):
        R = 0.0
        if not traj["done"]:
            s_tensor = torch.as_tensor(traj["next_state"], dtype=torch.float32)
            with torch.no_grad():
                _, next_val = self.model(s_tensor)
            R = next_val.squeeze(0).item()

        returns = []
        for r in reversed(traj["rewards"]):
            R = r + GAMMA * R
            returns.insert(0, R)
        returns = torch.tensor(returns)
        values_t = torch.stack(traj["values"])

        if returns.shape != values_t.shape:
            return torch.tensor(0.0)

        adv = returns - values_t
        p_loss = -(torch.stack(traj["log_probs"]) * adv.detach()).mean()
        v_loss = 0.5 * F.mse_loss(values_t, returns)
        return p_loss + v_loss

    def _apply_gradients(self, loss):
        self.opt.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(self.model.parameters(), 1.0)
        self.opt.step()
```

```python
class A3CTrainer:

    def __init__(self):
        env = gym.make(ENV_ID)
        obs_dim  = env.observation_space.shape[0]
        act_dim = env.action_space.n
        env.close()
        self.model = ActorCritic(obs_dim, act_dim)
        self.model.share_memory()
        self.opt = torch.optim.Adam(self.model.parameters(), lr=LR)
        self.counter = mp.Value('i', 0)
        self.queue = mp.Queue()

    def train(self):
        workers = []
        for i in range(NUM_WORKERS):
            w = A3CWorker(i, ENV_ID, self.model, self.opt, self.counter, self.queue)
            w.start()
            workers.append(w)
        rewards = []
        while len(rewards) < MAX_EPISODES:
            tag, val = self.queue.get()
            if tag == "[reward]":
                rewards.append(val)
                if len(rewards) % 10 == 0:
                    print(f"Episode Number: {len(rewards)} | Avg Reward For Last10 Eps. = {np.mean(rewards[-10:]):.2f}")
                if len(rewards) % 50 == 0 and np.mean(rewards[-50:]) >= 480:
                    with self.counter.get_lock():
                        self.counter.value = MAX_EPISODES
                    print("==== Environment Solved ====", "\n ==== Initiating Early Stopping ====")
                    break
        for w in workers:
            w.join()
        self._plot(rewards, f"A3C on {ENV_ID} ({NUM_WORKERS} workers)", SMOOTHING)
        return rewards, self.model

    def evaluate(self, model):
        env   = gym.make(ENV_ID)
        rews  = []
        model.eval()
        for i in range(EVAL_EPISODES):
            s, _ = env.reset()
            done  = False
            tot   = 0
            while not done:
                with torch.no_grad():
                    logits, _ = model(torch.as_tensor(s, dtype=torch.float32))
                a, s, r, term, trunc, _ = (torch.argmax(logits).item(),) + env.step(torch.argmax(logits).item())
                done = term or trunc
                tot += r
            rews.append(tot)
            print(f"Eval ep {i+1}: {tot}")
        env.close()
        self._plot(rews, "Greedy Evaluation", 1)
        print("Avg eval reward:", np.mean(rews))

    def _plot(self, rewards, title, smooth):
        plt.figure(figsize=(10,5))
        plt.plot(rewards, alpha=0.6, label="Reward")
        if len(rewards) >= smooth:
            ma = np.convolve(rewards, np.ones(smooth)/smooth, mode='valid')
            plt.plot(range(smooth-1, len(rewards)), ma, lw=2, label=f"{smooth}-ep mov avg")
        plt.xlabel("Episode")
        plt.ylabel("Total Reward")
        plt.title(title)
        plt.grid(True)
        plt.legend()
        plt.show()
```

```python
mp.set_start_method("fork")
trainer = A3CTrainer()
training_rewards, model = trainer.train()
```

```python
trainer.evaluate(model)
```
